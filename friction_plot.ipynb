```python
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
import csv
import glob
path = 'data' # use your folder name
file_name = glob.glob(path + "/*.csv")

t=[]
c=0
M_final=[]
K_final=[]

M1=[]
K1=[]
jnt_velocity=[]
friction_torque=[]
content = genfromtxt(file_name[0], delimiter=' ')
for i in content[:,0]:
    c=c+1
    t.append(c)
for i in content[:,2]:
    jnt_velocity.append(i)
for i in content[:,6]:
    friction_torque.append(i)
ty= np.round(jnt_velocity,3)
jv=[]
ft=[]
dic={}
for i in range(len(t)):
    if (i>2500):
        jv.append(ty[i])
        ft.append(friction_torque[i])
        dic[ty[i]]= friction_torque[i]
key=[]
value=[]

for k,v in dic.items():
    key.append(float(k))
    value.append(v)
for jl in range(len(key)):
    l=[i for i, e in enumerate(jv) if e == key[jl]]
    M1.append((np.mean([ft[jk] for jk in l])))
mul= int(180/np.pi)
for j in range(len(key)):
    K1.append(key[j]*mul)

for i in range(len(M1)):
    M_final.append(M1[i])
    K_final.append(K1[i])
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(5)
plt.scatter(K1, M1, color='b', label='Friction Torque')
plt.legend()
plt.grid()
plt.xlabel("Joint Velocity (deg/sec)")
plt.ylabel("[Nm]")
plt.savefig('plots/fr_tq_vs_jnt_vel_0_1.png')
plt.show()

M2=[]
K2=[]
c=0
t=[]
jnt_velocity=[]
friction_torque=[]
content = genfromtxt(file_name[1], delimiter=' ')
for i in content[:,0]:
    c=c+1
    t.append(c)
for i in content[:,2]:
    jnt_velocity.append(i)
for i in content[:,6]:
    friction_torque.append(i)
ty= np.round(jnt_velocity,3)
jv=[]
ft=[]
dic={}
for i in range(len(t)):
    if (i>1500):
        jv.append(ty[i])
        ft.append(friction_torque[i])
        dic[ty[i]]= friction_torque[i]
key=[]
value=[]

for k,v in dic.items():
    key.append(float(k))
    value.append(v)
for jl in range(len(key)):
    l=[i for i, e in enumerate(jv) if e == key[jl]]
    M2.append((np.mean([ft[jk] for jk in l])))
mul= int(180/np.pi)
for j in range(len(key)):
    K2.append(key[j]*mul)

for i in range(len(M2)):
    M_final.append(M2[i])
    K_final.append(K2[i])
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(5)
plt.scatter(K2, M2, color='g', label='Friction Torque')
plt.legend()
plt.grid()
plt.xlabel("Joint Velocity (deg/sec)")
plt.ylabel("[Nm]")
plt.savefig('plots/fr_tq_vs_jnt_vel_0_3.png')
plt.show()

M3=[]
K3=[]
jnt_velocity=[]
friction_torque=[]
c=0
t=[]
content = genfromtxt(file_name[2], delimiter=' ')
for i in content[:,0]:
    c=c+1
    t.append(c)
for i in content[:,2]:
    jnt_velocity.append(i)
for i in content[:,6]:
    friction_torque.append(i)
ty= np.round(jnt_velocity,3)
jv=[]
ft=[]
dic={}
for i in range(len(t)):
    if (i>7000):
        jv.append(ty[i])
        ft.append(friction_torque[i])
        dic[ty[i]]= friction_torque[i]
key=[]
value=[]

for k,v in dic.items():
    key.append(float(k))
    value.append(v)
for jl in range(len(key)):
    l=[i for i, e in enumerate(jv) if e == key[jl]]
    M3.append((np.mean([ft[jk] for jk in l])))
mul= int(180/np.pi)
for j in range(len(key)):
    K3.append(key[j]*mul)    
    
for i in range(len(M3)):
    M_final.append(M3[i])
    K_final.append(K3[i])   
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(5)
plt.scatter(K3, M3, color='r', label='Friction Torque')
plt.legend()
plt.grid()
plt.xlabel("Joint Velocity (deg/sec)")
plt.ylabel("[Nm]")
plt.savefig('plots/fr_tq_vs_jnt_vel_0_5.png')
plt.show()

M5=[]
K5=[]
jnt_velocity=[]
friction_torque=[]
c=0
t=[]
content = genfromtxt(file_name[4], delimiter=' ')
for i in content[:,0]:
    c=c+1
    t.append(c)
for i in content[:,2]:
    jnt_velocity.append(i)
for i in content[:,6]:
    friction_torque.append(i)
ty= np.round(jnt_velocity,3)
jv=[]
ft=[]
dic={}
for i in range(len(t)):
    if (i>3500):
        jv.append(ty[i])
        ft.append(friction_torque[i])
        dic[ty[i]]= friction_torque[i]
key=[]
value=[]

for k,v in dic.items():
    key.append(float(k))
    value.append(v)
for jl in range(len(key)):
    l=[i for i, e in enumerate(jv) if e == key[jl]]
    M5.append((np.mean([ft[jk] for jk in l])))
mul= int(180/np.pi)
for j in range(len(key)):
    K5.append(key[j]*mul)

for i in range(len(M5)):
    M_final.append(M5[i])
    K_final.append(K5[i])    
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(5)
plt.scatter(K5, M5, color='magenta', label='Friction Torque')
plt.legend()
plt.grid()
plt.xlabel("Joint Velocity (deg/sec)")
plt.ylabel("[Nm]")
plt.savefig('plots/fr_tq_vs_jnt_vel_1_3.png')
plt.show()

M4=[]
K4=[]
jnt_velocity=[]
friction_torque=[]
c=0
t=[]
content = genfromtxt(file_name[3], delimiter=' ')
for i in content[:,0]:
    c=c+1
    t.append(c)
for i in content[:,2]:
    jnt_velocity.append(i)
for i in content[:,6]:
    friction_torque.append(i)
ty= np.round(jnt_velocity,3)
jv=[]
ft=[]
dic={}
for i in range(len(t)):
    if (i>3500):
        jv.append(ty[i])
        ft.append(friction_torque[i])
        dic[ty[i]]= friction_torque[i]
key=[]
value=[]

for k,v in dic.items():
    key.append(float(k))
    value.append(v)
for jl in range(len(key)):
    l=[i for i, e in enumerate(jv) if e == key[jl]]
    M4.append((np.mean([ft[jk] for jk in l])))
mul= int(180/np.pi)
for j in range(len(key)):
    K4.append(key[j]*mul)        

for i in range(len(M4)):
    M_final.append(M4[i])
    K_final.append(K4[i])    
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(5)
plt.scatter(K4, M4, color='orange', label='Friction Torque')
plt.legend()
plt.grid()
plt.xlabel("Joint Velocity (deg/sec)")
plt.ylabel("[Nm]")
plt.savefig('plots/fr_tq_vs_jnt_vel_0_9.png')
plt.show()        
```
